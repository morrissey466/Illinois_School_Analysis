```python
import pandas as pd
import os
```

# Creating District dataframe

```python
# Creating dataframe for the General worksheet
general_df = pd.read_csv('General.csv')
general_df.head(10)
```

```python
# checking for na
general_df.isna().sum()
```

```python
# filtering for district
district_general_df = general_df.loc[general_df['Type'] == "District"]
```

```python
district_general_df.head(10)
```

```python
district_general_df.isna().sum()
```

```python
# dropping na with anything over 750
district_general_df = district_general_df.dropna(thresh=750, axis=1)
```

```python
# filtering for district
district_general_df = general_df.loc[general_df['Type'] == "District"]
district_general_df = general_df.loc[general_df['District Type'].isin(["UNIT","HIGH SCHOOL"])]
```

```python
list(district_general_df.columns)
```

```python
district_general_df.head()
```

```python
# The columns that are chosen for the District dataframe
district_filtered_df = district_general_df[["RCDTS", 
    "Type",  
    "District", 
    "City", 
    "County",
    "District Type", 
    "District Size", 
    "Student Enrollment - Total",
    "Student Enrollment - Low Income %",
    "Avg Class Size – All Grades",
    "Avg Teaching Exp",
    "Bachelor Degree",
    "Masters Degree",
    "Teacher Retention Rate",
    "Teacher Avg Salary",
    "Admin Avg Salary"]]
district_filtered_df.head(20)
```

```python
district_filtered_df = district_filtered_df.dropna()
```

```python
district_filtered_df.count()
```

```python
district_filtered_df['RCDTS']= '0' + district_filtered_df['RCDTS'].astype(str)
district_filtered_df.head(20)
```

```python

# Exporting dataframe to clean data folder.
district_filtered_df.to_csv('clean_data/district_filtered.csv', index=False)
```

# Creating High School Dataframe from general csv

```python
general_df.head()
```

```python
# filtering for district
high_school_general_df = general_df.loc[general_df['School Type'] == "HIGH SCHOOL"]
high_school_general_df.head()
```

```python
high_school_general_df.isna().sum()
```

```python
# dropping na with anything over 750
high_school_general_df = high_school_general_df.dropna(thresh=100, axis=1)
high_school_general_df.head(20)
```

```python
list(high_school_general_df.columns)
```

```python
high_school_general_df.isna().sum()
```

```python
high_school_filtered_df = high_school_general_df[["RCDTS", 
    "Type",  
    "District", 
    "School Type",
    "City", 
    "County",
    "District Type", 
    "District Size", 
    "Student Enrollment - Total",
    "Student Enrollment - Low Income %",
    "Avg Class Size – All Grades",
    "Teacher Retention Rate",
    "# of 12 graders who took advanced Courses",
    'Avg Number of days of Physical Education Per Week Per Student',
    'Community College Remediation %',
    'Avg Class Size - High School',
    "Chronic Absenteeism",
    'Number of students who took AP Exams Grade 10',
    'Number of students who passed AP Exams Grade 12',
    "Chronic Absenteeism - Low Income"]]
high_school_filtered_df.head(20)
```

```python
high_school_filtered_df['Low Income Count']= high_school_filtered_df['Student Enrollment - Total']*(high_school_filtered_df['Student Enrollment - Low Income %']/100)
high_school_filtered_df.head(20)
```

```python
high_school_filtered_df.count()
```

```python
high_school_filtered_df = high_school_filtered_df.dropna()
```

```python
high_school_filtered_df['RCDTS']= '0' + high_school_filtered_df['RCDTS'].astype(str)
high_school_filtered_df.head(20)
```

```python
high_school_filtered_df.count()
```

```python
high_school_filtered_df.head(30)
```

```python
# Exporting dataframe to clean data folder.
high_school_filtered_df.to_csv('clean_data/high_school_filtered.csv', index=False)
```
